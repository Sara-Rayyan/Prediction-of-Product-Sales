# creating exploratory visuals

# Histograms to view the distributions of numerical features in your dataset.

````python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(10,6))
    ax = sns.histplot(df[col], bins=30)
    ax.set(title = f"Distribution of {col}")
    ax.grid(linestyle = ':');
    plt.show()
    print('\n')
````

# Boxplots to view statistical summaries of numerical features in your dataset.

- **Features are:**
1.   Item_Outlet_Sales
2.   Outlet_Type
---
- **Why these two columns?**
```
Item_Outlet_Sales:
```
This is the target variable (what we want to understand and eventually predict). Analyzing its distribution is crucial.

```
Outlet_Type:
```
A categorical feature that represents the type of store (grocery vs. supermarket variations). Comparing sales across these categories helps identify which store types are more profitable and whether this feature can be an important predictor in modeling.

---
- **Reason for choosing these columns together:**
> It is crucial to understand how sales performance varies across different types of outlets. Sales (Item_Outlet_Sales) represent the core revenue driver, while the outlet type (Outlet_Type) reflects the business model and customer reach of each store format (e.g., grocery store vs. supermarket). Comparing these two variables helps identify which outlet formats generate higher sales, highlight potential inefficiencies, and guide strategic decisions such as where to launch promotions, introduce new products, or expand operations.





````python
plt.figure(figsize=(10,6))
ax = sns.boxplot(data = df, x = 'Item_Outlet_Sales', hue = 'Outlet_Type');
ax.legend(loc = 0);
plt.tight_layout()
````

- **What the Boxplots shows?**

> The boxplot reveals that Supermarket Type3 achieves the highest median sales, indicating it typically has better performance than other store types. In contrast, Grocery Stores show the lowest median sales, suggesting they generally underperform compared to supermarket formats.

> Additionally, there is a relative convergence in sales between Supermarket Type1 and Supermarket Type2, though Type1 maintains a slightly higher median.

> Notably, Supermarket Type1 displays the largest number of outliers on the higher end of the sales distribution. This could be due to several factors, such as:

* Locations in high-traffic or high-income areas
* Effective promotional strategies or larger discounts
* A wider product assortment or premium, higher-priced products







# Countplots to view the frequency of each class of categorial features in your dataset.

- **Features are:**
1.   Item_Type (The category to which the product belongs)
2.   Item_Fat_Content(Whether the product is low-fat or regular)
---
- **Reason for choosing these columns together:**

> We selected Item_Type and Item_Fat_Content to understand consumer preferences and product distribution. By analyzing which product types are more likely to be low-fat or regular, retailers can gain actionable insights for inventory planning, targeted promotions, and optimizing product placement to better meet customer demand.










````python
plt.figure(figsize=(12,6))
ax = sns.countplot(data=df, x='Item_Type', hue='Item_Fat_Content')
plt.xticks(rotation=90)
plt.title("Distribution of Product Types by Fat Content")
ax.legend(loc = 1)
plt.show()
````

- **What the countplot shows?**


> Consumers show a clear preference for Low Fat options in categories like Snack Foods and Soft Drinks, while Regular products dominate in Meat and Breakfast. Other categories, such as Fruits & Vegetables and Frozen Foods, are relatively balanced, highlighting opportunities to expand healthier product lines.
* Snack Foods & Soft Drinks → Strong shift toward Low Fat options
* Meat & Breakfast → Regular products remain dominant
* Fruits & Vegetables, Frozen Foods → Balanced mix, growth potential for Low Fat
* Smaller categories (Seafood, Starchy Foods, Others) → Limited market impact
